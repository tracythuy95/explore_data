```python
import pandas as pd
import seaborn as sns
import altair as alt
import numpy as np
import pickle
```

```python
df = pickle.load( open( "20210125_leggings_hoodies_shorts_tops", "rb" ) )
df
```

```python
boxplot = alt.Chart(df).mark_boxplot().encode(
    x='material_clean_list',
    y='avg_rating',
    tooltip=['product_name', 'current_price', 'old_price', 'type']
).properties(
    width=700,
    height=300
).interactive()

boxplot
```

```python
brush = alt.selection(type='interval')

points = alt.Chart(df).mark_point().encode(
    x='old_price',
    y='current_price',
    color=alt.condition(brush, 'total_reviews', alt.value('lightgray')),
    tooltip=['product_name', 'current_price', 'old_price', 'type', 'avg_rating']
).add_selection(
    brush
).properties(
    width=700,
    height=300
).interactive()

bars = alt.Chart(df).mark_bar().encode(
    y='type',
    color='type',
    x='count(type)'
).transform_filter(
    brush
).properties(
    width=800,
    height=200
)

bars2 = alt.Chart(df).mark_bar().encode(
    y='material_clean_list',
    color='material_clean_list',
    x='count(material_clean_list)'
).transform_filter(
    brush
).properties(
    width=800,
    height=600
)

points & bars & bars2
```
